<h2>1. Définition du problème</h2>

Pour ce projet, il nous a été demandé de proposer une solution au probléme de tournées de véhicules. Dans notre contexte, une entreprise veut investir dans l'optimisation de son service de livraison. Nous allons devoir definir une carte, qui sera composées de plusieurs villes à parcourir par les livreurs. L'objectif est de parcourir la carte en passant par chaque ville une seule fois sans possibilitée d'y repasser durant la tournée. Le but étant de renvoyer le livreur à son point de départ, et de minimiser la distance de la tournée ainsi que la durée de livraison.

<h3>1.2 Identification du problème au problème de Tournée de véhicules </h3>


Le problème du voyageur de commerce est un cas particulier du problème VRP sans contrainte de capacité ; il consiste en la détermination d'un parcours de coût minimal (distance, temps, etc.) pour un seul véhicule partant d'une localité, visitant n autres localités et revenant à son point de départ.

<h2>2. Définition mathématique du problème</h2>

Un Programme Mathématique (mathematical program), noté PM, est un problème d'optimisation sous contrainte (P) qui peut s'écrire de la façon suivante :

    Maximiser f(x)
    sous les contraintes
    gi(x) ≤ 0 i = 1, ..., m
    x ∈ S.
    
Rappelons les contraintes choisies :

$n$ nombre de clients (ou sommets)<br>
$m$ nombre de véhicules<br>
$Q$ capacités des véhicules<br>
$q_{i}$ demande des clients<br>
$c_{i,j}$ Poids de l'arete entre $i$ et $j$<br>

Les variables de décision du problème sont les $x_{ijk}$  évoquées plus haut : <br>

$x_{ijk}$ = 1 si $(i,j)$ est parcouru par le véhicule $k$, si non 0<br>

Ainsi, eb tant que problème d'optimissation, le CVRP s'écrit : <br>
$$\min \sum_{i=1} ^n \sum_{j=1} ^m c_{ij}  \sum_{k=1} ^m x_{ijk} $$ 

<br>
sujet aux contraintes suivantes :<br>
$$ \sum_{i=1} ^n \sum_{k=1} ^m x_{ijk} = 1 \quad \forall \ 1 \leq j \leq n$$

$$ \sum_{j=1} ^n \sum_{k=1} ^m x_{ijk} = 1 \quad \forall \ 1 \leq i \leq n$$

$$\sum_{i=1} ^n \sum_{l=1} ^n x_{ilk} =  \sum_{l=1} ^n \sum_{j=1} ^n x_{ljk} $$

$$\sum_{j=1} ^n x_{0jk} = 1\quad \forall \ 1 \leq k \leq m$$

$$\sum_{i=1} ^n x_{i0k} = 1\quad \forall \ 1 \leq k \leq m$$

$$ \sum_{i=1} ^n \sum_{j=1} ^m x_{ijk} \leq Q \quad\forall1 \leq k \leq m$$

$$x_{ijk} \in 0,1 \quad\forall 0 \leq i,j \leq n; 1 \leq k \leq m$$


<h2>3. Complexité </h2>


<h3>3.1 La complexité du problème de décision est expliquée (calcul ou référence scientifique).</h3>



Un problème de recherche peut parfois être transformé en un problème de décision équivalent. Par exemple, le problème du voyageur de commerce qui cherche, dans un graphe dont les arêtes sont étiquetées par des coûts, à trouver un cycle de coût minimum, passant une fois par chaque sommet, peut s'énoncer en un problème de décision comme suit : Existe-t-il un cycle hamiltonien (passant une fois par chaque sommet) de coût inférieur à k (donné dans l'instance) ? L'équivalence de ces deux problèmes doit être démontrée, elle montre que la démonstration d'existence repose sur un argument constructif.

<h3>3.2 L'étude de la complexité est adaptée au problème (Tournée de véhicules ou Voyageur du commerce)</h3>

<h2>6. Etude Statistique :</h2>

### 6.1 Introduction :

Dans cette partie-là, nous allons nous intéresser à la notion de statistique afin de faire une analyse prédictive du trafic routier. Pour cela, nous allons utiliser les statistiques prédictives, plus particulièrement les tests d'hypothèse et la régression linéaire.

1. <U>Petits rappels</U> :

    * <strong>Statistique</strong> : La statistique a pour objet de recueillir des observations portant sur des sujets présentant une certaine propriété et de traduire ces observations par des nombres qui permettent d’avoir des renseignements sur cette propriété.
    <br><br>
    * <strong>Individu</strong> : C’est un élément individuel (objet) appartenant à la population de notre étude.
    <br><br>
    * <strong>Population</strong> :  Constitue l'ensemble de tous les individus considérés dans notre analyse statistique.
    <br><br>
    * <strong>Echantillon</strong> :  Ensemble d'individus représentatifs d'une population. L'échantillonnage vise à obtenir une meilleure connaissance d'une ou plusieurs population(s) ou sous-populations(s) par l'étude d'un nombre d'échantillons jugé statistiquement représentatif.
    <br><br>
    * <strong>Statistique descriptive</strong> :  L'objectif de la statistique descriptive est de décrire, c'est-à-dire de résumer ou représenter, par des statistiques, les données disponibles quand elles sont nombreuses.
    <br><br>
    * <strong>Modèle prédictif</strong> :  Ce type de modèle permet d’analyser des faits présents et passés pour faire des hypothèses prédictives sur des événements futurs afin d'améliorer la notion d’efficacité.
    <br><br>
    * <strong>Régression linéaire</strong> :  C’est un modèle qui cherche à établir une relation linéaire entre une variable X et une variable Y. l’idée est de pouvoir ensuite faire des prévisions sur Y lorsque X est mesurée. Dans notre cas, la question qu’on se pose est la suivante : « Existe-t-il une relation linéaire entre le temps X (moments dans la journée) et le trafic routier Y ? ».
    <br><br>
    * <strong>Méthode des moindres carrés</strong> :  permet de comparer des données expérimentales, généralement entachées d’erreurs de mesure, à un modèle mathématique censé décrire ces données.
    <br><br>
    * <strong>Variance</strong> :  Permet d'étudier les variations d'une variable par rapport à elle-même.
    <br><br>
    * <strong>Covariance</strong> :  Permet d'étudier les variations simultanées de deux variables par rapport à leur moyenne respective.
    <br><br>
    * <strong>Ecart type</strong> :  Représente une mesure de la dispersion des valeurs d'un échantillon statistique ou d'une distribution de probabilité. Il est défini comme la racine carrée de notre variance.
    <br><br>
    * <strong>Tests statistiques</strong> :  Est une procédure de décision entre deux hypothèses. Il s'agit d'une démarche consistant à rejeter ou à ne pas rejeter une hypothèse statistique, appelée hypothèse nulle, en fonction d'un jeu de données (échantillon).
    <br><br>
    * <strong>Hypothèse statistique</strong> :  Est une explication anticipée, une affirmation provisoire qui décrit ou explique un phénomène. Elle est une prédiction consistant à mettre en relation une variable et un comportement.

### 6.3	Méthode des moindres carrées :

Pour cela, nous allons utiliser ce qu’on appelle la « méthode des moindres carrées ». Cette méthode va nous permettre de minimiser la somme des écarts (ei) que nous avons entre nos points et la droite que nous avons tracé. Ceci va donc nous donner l’équation de la droite qui passe au mieux par les points.
Il faudrait donc connaitre les valeurs des deux paramètres « a » et « b ». D’après la méthode des dérivées partielles (par rapport aux deux paramètres), nous obtenons les équations suivantes : 

$a$ = $\frac{COV(X;Y)}{V(X)}$

<br>

$b$ = $\bar{y} - a\bar{x}$

Avec : <br>

 * $COV(X;Y)$: Convariance de $X$ et $Y$
 * $V(X)$: Ou VAR(x) variance de $X$

<br>
<br>
ce qui nous donne : <br>

$a$ = $\frac{\bar{xy}-\bar{x}.\bar{y}}{\bar{x^2}-\bar{x}^2}$
<br>
$b =$ $\bar{y} - a.\bar{x}$
<br>

### 6.4	Mesure de la qualité de l’ajustement (performance) :
Ici nous allons utiliser le coefficient de corrélation linéaire afin de mesurer la qualité de notre prédiction. Il va donc nous donner une idée sur la droite obtenue et si elle est finalement adaptée aux valeurs (individus) que l’on a. Il permet de détecter la présence ou l'absence d'une relation linéaire entre les deux caractères quantitatifs continus (temps et trafic).
Comme notre cas correspond à une fonction affine, ce qui est un cas particulier, nous avons l’équation suivante :

<br>

$r$ = $\frac{COV(X;Y)}{\sigma(X). \sigma(Y)}$ <br>

Avec: <br>
 * COV(X,Y): covariance de $X$ et de $Y$ (expliqué précédemment)
 * $\sigma(X)$: éract type de $X$
 * $\sigma(Y)$: éract type de $Y$<br>
 
 
Ce qui nous donne : <br>

$r$ = $\frac{COV(X;Y)}{\sqrt{V(X)}. \sqrt{V(Y)}}$ <br>

<strong><u>NB :</u></strong>
<br>

Le coefficient de corrélation varie entre [0, 1]. Lorsqu'il est proche de 0, le pouvoir prédictif du modèle est faible et lorsqu'il est proche de 1, le pouvoir prédictif du modèle est fort. En revanche, s’il est égal à 0, on dit que nous avons une absence de corrélation (valeurs très dispensées), et s’il est égale à 1 on dit que notre corrélation est parfaitement positive (tous les points suivent la droite)

<br>
Dans notre cas, notre corrélation vaut :
<br>
<h4 style="color: red">$r \simeq 0.20$</h4> <br>
La valeur obtenue est inférieure à 0.5. Ce qui constitue une corrélation de faible intensité.
En ce qui concerne le coefficient de détermination, quant à lui, c’est un indicateur qui permet de juger la qualité de notre régression linéaire. Pour le calculer, il suffit de mettre au carré notre coefficient de corrélation :

<h4>$R = r^2$</h4> <br>
Ce qui nous donne : <br>
<h4 style="color: green">$R = 0.041$</h4><br>
On sait que plus le coefficient de détermination se rapproche de 0, plus le nuage de points se disperse autour de la droite de régression. Au contraire, plus il tend vers 1, plus le nuage de points se resserre autour de la droite de régression. Dans notre cas on peut conclure que les points sont dispersés autour de notre droite. <br>


### 6.5	Tests statistiques : 

Les tests statistiques sont des outils d’aide à la décision, qui permettent de vérifier des hypothèses expérimentales, avec un certain seuil de probabilité. Ils permettent, entre autres, de vérifier s’il y a une réelle différence entre les données observées et les données de la population, ou bien, la différence est dû simplement à une fluctuation d’échantillonnage.

<br>
Généralement, on pose deux hypothèses statistiques H0 et H1 :

 * -<u>H0 ou hypothèse nulle :</u> est souvent définie comme une absence de différence. Autrement dit, les paramètres de la population source et ceux de l’échantillon sont identiques, donc que la fluctuation observée provient de l’échantillonnage.
<br>

 * -<u>H1 ou hypothèse alternative :</u> correspond à la négation de H0, et donc dans ce cas, on observe une différence entre les paramètres de la population source et ceux de l’échantillon. Sous cette hypothèse, les fluctuations de l’échantillonnage ne suffisent pas à expliquer la différence observée entre les paramètres étudiés ;
 <br>
 
Après avoir défini nos hypothèses, on suppose que l’hypothèse H0 est vraie.<br>

Ensuite, il faut effectuer ce qu’on appelle un « test de signification de la pente ». Ce dernier permet de décider s’il existe un lien significatif, autrement dit si la corrélation observée entre X et Y existe bel et bien au sein de notre population. Ce teste peut se faire sur Excel, comme il peut se faire en ligne.<br>
Dans notre cas, on prend les tests effectués en ligne, ce qui nous donne :<br>



<strong><u>NB :</u></strong> <br>

la valeur-p (p-value) est la probabilité pour un modèle statistique donné sous l'hypothèse nulle d'obtenir la même valeur ou une valeur encore plus extrême que celle observée.
Comme on peut le constater, la valeur de notre p-value est très inférieure au seul significatif (qui est de 5%) :<br>

$p - value = 1,061.10-8 \lll 0.05$<br>
                                     

Sachant que : <br>

 * Si la p-value est supérieure à 0,05, on accepte l'hypothèse nulle.
 * Si la p-value est inférieure à 0,05, on rejette l'hypothèse nulle.<br>
 
Dans notre cas, on rejette l’hypothèse H0. On peut donc dire qu’il existe bel et bien une corrélation entre X et Y au sein de notre population.<br>

<strong><u>Remarque :</u></strong> <br>

On utilise des tests statistiques pour trois raisons différentes, soit : <br>

 * Pour comparer la moyenne observée et la moyenne théorique ;
 * Pour comparer deux variances (en utilisant le test de Fisher) ;
 * Pour comparer deux moyennes observées ;

Il existe de nombreux tests statistiques classiques parmi lesquels on peut citer : <br>


 - <u>Le test de Student</u> : qui sert à la comparaison d'une moyenne observée avec une valeur « attendue » pour un échantillon distribué selon une loi normale (on utilise ce test si on ne connaît pas la vraie variance de la population et surtout sur une petite population (avec n < 30 observations) ;

 - <u>Le test du Fisher</u> : qui sert à la comparaison de deux variances observées ;
 
 - <u>Le test 7</u> : permet de comparer les moyennes de deux échantillons si on connaît la vraie variance σ² de la population (avec n supérieur ou égale à 30 observations).
